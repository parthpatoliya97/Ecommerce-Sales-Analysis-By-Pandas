```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,6))

plt.subplot(2,2,1)
plt.title("Top Left")
plt.plot([1,2,3],[1,4,9])

plt.subplot(2,2,2)
plt.title("Top Right")
plt.plot([1,2,3],[1,2,3])

plt.subplot(2,2,3)
plt.title("Bottom Left")
plt.plot([1,2,3],[3,2,1])

plt.subplot(2,2,4)
plt.title("Bottom Right")
plt.plot([1,2,3],[9,4,1])

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('IPL.csv')
df.head(5)
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
match_wins=df['match_winner'].value_counts()
match_wins
```

```python
sns.barplot(y=match_wins.index,x=match_wins.values)
plt.title('Most match win by Team')
```

```python
sns.countplot(x=df['toss_decision'])
plt.title('Toss decision trends')
```

```python
df[df['toss_winner']==df['match_winner']]
```

```python
twmw = df[df['toss_winner'] == df['match_winner']]['match_id'].count()
percentage = (twmw * 100) / df.shape[0]
print("Percentage of matches where toss winner also won:", percentage.round(2), "%")
```

```python
sns.countplot(x=df['won_by'])
```

```python
potm_count=df['player_of_the_match'].value_counts()
potm_count
```

```python
potm_count=df['player_of_the_match'].value_counts().head(10)
potm_count
```

```python
sns.barplot(x=potm_count.values,y=potm_count.index)
plt.title('Top 10 POTM Players')
```

```python
top_score=df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(3)
top_score
```

```python
top_score.plot(kind='barh')
```

```python
df['highest_wickets']=df['best_bowling_figure'].apply(lambda x:x.split('--')[0])
df['highest_wickets']=df['highest_wickets'].astype(int)
```

```python
top_bowlers=df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending=False).head(10)
top_bowlers
```

```python
top_bowlers.plot(kind='barh')
```

```python
venue_count=df['venue'].value_counts()
venue_count
```

```python
sns.barplot(x=venue_count.values,y=venue_count.index)
```

```python
df[df['won_by']=='Runs'].sort_values(by='margin',ascending=False).head(1)[['match_winner','margin']]
```

```python
df[df['highscore']==df['highscore'].max()][['top_scorer','highscore']]
```

```python
df[df['highest_wickets']==df['highest_wickets'].max()][['best_bowling','best_bowling_figure']]
```
